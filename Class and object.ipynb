```python
#praktikum 1
#No 1
def waktu():
    import time
    a=time.localtime()
    hr=a.tm_hour
    mn=a.tm_min
    sc=a.tm_sec
    return ('{}:{}:{}'.format(hr,mn,sc))
waktu()

def ucapan():
    nama = input("Masukkan Nama: ")
    a=waktu()
    if a > "00:00:00" and a < "10:00:00":
        print("Selamat Pagi", nama)
    elif a > "10:00:00" and a < "15:00:00":
        print("Selamat siang", nama)
    elif a > "15:00:00" and a < "18:00:00":
        print("Selamat Sore", nama)
    else :
        print ("Selamat Malem", nama)
ucapan()

#No 2
#2A
a = 3
b = 7
c = 2
for i in range(a,b+1):
    for j in range(c):
        print (a,end="")
    a+=1
    print ()

#No. 2B
a = 8
b = 10
c = 4
for i in range(a,b+1):
    for j in range(c):
        print (a,end="")
    a+=1
    print ()

#3
#No. 3A dan 3B
for i in range(2):
    a = int(input("a : "))
    b = int(input("b : "))
    for i in range(a, b+1, 2):
        #for j in range (a):
        print(i)
#No. 3C
a = int(input("a : "))
b = int(input("b : "))
for i in range(a+1, b, 2):
    #for j in range (a):
    print(i)
```

```python
import datetime
now = datetime.datetime.now()

class pegawai :
    def __init__ (self,No,nm,tgllahir) :
        self.Nomer=No
        self.Nama_Pegawai=nm
        self.Tanggal_Lahir=tgllahir
    
    def tampilkan(self):
        lahir = self.Tanggal_Lahir
        tahunLahir = int(lahir[-4:])
        tahunSekarang = now.year
        umur = tahunSekarang - tahunLahir
        print("No. Pegawai :", self.Nomer, "Nama Pegawai :", self.Nama_Pegawai, "Umur :", umur)
        
        

fera=pegawai("170411100087", "Ana Maghfiroh", "12 maret 1999")
fera.tampilkan()
fera.umur()
```
